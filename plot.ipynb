```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/diabetes (1).csv')
```

```python
plt.figure(figsize=(5,5))
sns.boxplot(data=df)
plt.title('edit tatu vidu')
plt.xlabel('bala')
plt.ylabel('gay')
plt.show()
```
